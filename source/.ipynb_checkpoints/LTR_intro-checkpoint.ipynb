# 1. Introduction to LTR(Learning to rank)
LTR はLearning To Rank の頭文字です.   
日本では,ランキング学習やランク学習と呼ばれています．

## 1.1 LTR(ランキング学習)って?
LTRはランク付け問題を解くために機械学習の技術を適用して問題を求解する機械学習の一つである．   
様々なものに適用でき， 例として, 単純なランキングを求める問題(競馬など)や, リコメンドシステムなどがある．
